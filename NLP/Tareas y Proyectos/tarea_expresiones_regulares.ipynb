```python
#Importamos las librerias que vamos a ocupar
import re
from dateutil import parser
```

```python
#Leemos el archivo
with open("dates.txt", "r") as myfile:
    string=myfile.read().replace('\n','') 
```

```python
#Creamos las expresiones regulares
a = re.compile(r'(\d{1,2}[\/-]\d{1,2}[\/-]\d{2,4})|([0-9]{1,2}\/[0-9]{2,4})|((\d{2} )*(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4})|((Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (\d{2})?, \d{4})|((19|20)\d{2})')
#La primer expresión regular es para las fechas tipo 03/25/93               \d{1,2}[\/-]\d{1,2}[\/-]\d{2,4})
#La segunda expresión regular es para las fechas de tipo 08/1970 o 08/70    ([0-9]{1,2}\/[0-9]{2,4})  
#La tercera expresión regular es para las fechas tipo 12 August 2012        ((\d{2} )*(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4})
#La cuarta expresión regular es para las fechas August 12, 2001             ((Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (\d{2})?, \d{4})
#La quinta expresión regular es para las fechas de tipo 2001                ((19|20)\d{2})

fechas = a.findall(string)
print(fechas)
```

```python
#Obtenemos las partes que nos intersan de las fechas ya que los elementos de fechas tienen otros formatos
fechas_ = []
for j, v in enumerate(fechas):
    a = 0
    b = 0
    for j1,v1 in enumerate(v):
        if a < len(v1):
            a = len(v1)   
            b = j1
    fechas_.append(fechas[j][b])

print(fechas_)
```
