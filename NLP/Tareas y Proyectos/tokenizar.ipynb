```python
ce1 = ['“','”']
```

```python
def tokenizador(texto):
    texto_ = texto
    a = 0
    for j,v in enumerate(texto):
        if v in ce1:
            texto_ = texto_[:j+a+1] + ' ' + texto_[j+a+1:]
            a += 1
        elif v  == ',':
            texto_ = texto_[:j+a] + ' ' + texto_[j+a:]
            a += 1
        elif v == '.':
            texto_ = texto_[:j+a] + ' ' + texto_[j+a:j+a+1] + ' ' + texto_[j+a+1:]
            a += 2
        elif v  == '!':
            texto_ = texto_[:j+a] + ' ' + texto_[j+a:]
            a += 1
        elif v  == '¡':
            texto_ = texto_[:j+a+1] + ' ' + texto_[j+a+1:]
            a += 1

    s1 = texto_.split()
    s1_ = []
    s1__ = []
    for j in s1:
        if j in s1_:
            s1__[s1_.index(j)] += 1

        else:
            s1_.append(j)
            s1__.append(1)

    x = [[s1_[j],s1__[j]] for j in range(len(s1_))]
    df1 = pd.DataFrame(x, columns=['type','repeticiones'])
    print('El número de tokens es', len(s1))
    print('El número de types es', len(s1_))
    return df1
```

```python
tokenizador(string1)
```
